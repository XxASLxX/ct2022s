# 作業程式碼

```python


allFileList = os.listdir("docs/長恨歌")

for file in allFileList:

  svg = SVG.load_svg("docs/長恨歌/" + file).normalize().zoom(0.9).canonicalize().simplify_heuristic()

  svg_target = SVGTensor.from_data(svg.to_tensor())
  p_target = svg_target.sample_points()
  #plot_points(p_target, show_color=True)

  shape = SVG.unit_triangle().normalize().zoom(0.9).split(9) 
  svg_pred = SVGTensor.from_data(shape.to_tensor())

  p_pred = svg_pred.sample_points()
  #plot_points(p_pred, show_color=True)

  svg_pred.control1.requires_grad_(True)
  svg_pred.control2.requires_grad_(True)
  svg_pred.end_pos.requires_grad_(True);

  optimizer = optim.Adam([svg_pred.control1, svg_pred.control2, svg_pred.end_pos], lr=1)

  img_list = []

  for i in range(250):
      optimizer.zero_grad()

      p_pred = svg_pred.sample_points()
      l = svg_emd_loss(p_pred, p_target)
      l.backward()
      optimizer.step()
      
      if i % 249 == 0:
          img = svg_pred.draw(with_points=True, do_display=False, return_png=True)
          svgtemp = SVG.from_tensor(svg_pred.data)
          svgtemp.save_svg("docs/HW7_result/" + file)

  print('docs/HW7_result/' + file)
  f = open('docs/HW7_result/' + file ,encoding="utf-8")
  line = f.readline()
  svg2png(bytestring = line, write_to= 'docs/HW7_result/png/U+'+ file[2:-4] +'.png')
  f.close()
```
